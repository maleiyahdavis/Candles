```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Here I am initializing my dataframe and reading the csv file for my dataset 
nba_injuries_df= pd.read_csv('injuries_2010-2020.csv')

#.head(50) will show me the first 50 rows of my dataframe (this is not a copy)
nba_injuries_df.head(50)
```

```python
#Here I am using .info() to see the breakdown of the dataframe 
nba_injuries_df.info()
```

```python
#Here I created a variable named values and set the key value pairs of Acquired and Relinquished to N/A
#I did this in order to use df.fillna to fill in the NAN values in those two columns
#I set value fill in to the variable values that I created 
#I also made a copy of this newly modified dataframe and named it values_df
values = {"Acquired": "N/A", "Relinquished": "N/A"}
values_df = nba_injuries_df.fillna(value= values).copy()
```

```python
team_occurance1 = values_df['Team'].value_counts()['Lakers']
team_occurance2 = values_df['Team'].value_counts()['Cavaliers']
team_occurance3 = values_df['Team'].value_counts()['Celtics']
team_occurance4= values_df['Team'].value_counts()['Rockets']

data = {
    "Team": ["Lakers", "Cavaliers", "Celtics", "Rockets"], 
    "Injuries": [team_occurance1, team_occurance2, team_occurance3, team_occurance4]
}
total_df = pd.DataFrame(data)
total_df
#values_df.groupby("Team").agg({"Team":[min, max]})
```

```python
total_df.Injuries.agg([min, max, sum])
```

```python
plt.figure(figsize=(8,8), dpi=150)

total_df.plot(x = "Team", y= "Injuries", kind= "bar")

plt.xlabel("Team")
plt.ylabel("# of Injuries")
plt.title("Number of Injuries In NBA 2010-2020 Seasons")

plt.xticks(rotation = 40)
plt.savefig("./NBAInjuries.jpg")
plt.show()
```

```python
#Here I am taking the values_df and using loc in order to obtain the Team key where it is equivalent to Cavaliers
#I making a copy of this dataframe and setting it equal to cavs_df 
cavs_df = values_df.loc[values_df.Team == "Cavaliers"].copy()
cavs_df.head(5)
```

```python
#Here I am creating copy of the previous dataframe and modifying the columns to Date, Team, Acquired, Relinquished
#Next I am setting the index of this newly created dataframe to Date
new_cavs_df = cavs_df[["Date", "Team", "Acquired", "Relinquished"]].copy()
new_cavs_df.set_index("Date")
```

```python
#Here I am taking the values_df and using loc in order to obtain the Team key where it is equivalent to Lakers
#I making a copy of this dataframe and setting it equal to lakers_df 
lakers_df = values_df.loc[values_df.Team == "Lakers"].copy()
lakers_df.head(5)
```

```python
#Here I am creating copy of the previous dataframe and modifying the columns to Date, Team, Acquired, Relinquished
#Next I am setting the index of this newly created dataframe to Date
new_lakers_df = lakers_df[["Date", "Team", "Acquired", "Relinquished"]].copy()
new_lakers_df.set_index("Date")
```

```python
c_no_new_player = new_cavs_df['Acquired'].value_counts()['N/A'] #654
l_no_new_player = new_lakers_df['Acquired'].value_counts()['N/A'] #660

plt.figure(figsize=(8,8), dpi=150)

x1 = ["Cavaliers"]
y1 = c_no_new_player
plt.bar(x1,y1, width= .08, color= "red")

x2= ["Lakers"]
y2= l_no_new_player
plt.bar(x2,y2, width= .08, color= "gold")

plt.xlabel("Team")
plt.ylabel("# of Players not Acquired")
plt.title("Number of Players Not Acquired After Injury: Caveliers and Lakers 2010-2020 Season")
plt.savefig("./NonAcquiredPlayers.jpg")
plt.show()
```
